```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/illiyarosida/datakb/main/kbfix.csv')
df.head(38)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df.KabupatenKota)
df.KabupatenKota = le.transform(df.KabupatenKota)
df.KabupatenKota
```

```python
X = df.drop(columns='KabupatenKota')
y = df.KabupatenKota
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df)
# kmeans.fit(X_train, y_train)
```

```python
yclus=kmeans.labels_
```

```python
print(kmeans.cluster_centers_)
```

### Modelling menghitung akurasi

```python
from sklearn.metrics import accuracy_score
accuracy_score(yclus, yclus)
```
